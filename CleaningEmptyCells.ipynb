## Cleaning Empty Cells

```python
import pandas as pd
```

```python
df=pd.read_csv('dirty_employee_data.csv')
print(df.to_string())
```

#### Remove Rows 

```python
new_df=df.dropna() #Remove rows with missing values in new created new table
print(new_df.to_string())
```

```python
df=pd.read_csv('dirty_employee_data.csv')  ##Remove rows with missing values in original table
df.dropna(inplace=True)
print(df.to_string())

```

#### Replace null value with 999

```python
df=pd.read_csv('dirty_employee_data.csv')
df.fillna(999,inplace=True)
print(df.to_string())

```

#### Replace 108 only in specified column

```python
df=pd.read_csv('dirty_employee_data.csv')
df["Age"]=df["Age"].fillna(108)
print(df.to_string())
```

#### Replace using mean, median or mode

```python
df=pd.read_csv('dirty_employee_data.csv')
x=df['Age'].mean()
df['Age']=df['Age'].fillna(x)
print(df.to_string())

```

```python
df=pd.read_csv('dirty_employee_data.csv')
x=df['Age'].median()
df['Age']=df['Age'].fillna(x)
print(df.to_string())

```

```python
df=pd.read_csv('dirty_employee_data.csv')
x=df['Age'].mode()[0]
df['Age']=df['Age'].fillna(x)
print(df.to_string())

```
