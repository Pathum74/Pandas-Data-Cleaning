## Pandas Removing Duplicate Values

```python
import pandas as pd
```

```python
df=pd.read_csv('dirty_employee_data.csv')
print(df.to_string())
```

#### Find duplicate values

```python
print(df.duplicated())
```

#### Remove duplicate values

```python
df.drop_duplicates(inplace=True)
print(df.to_string())
```
